```
import pandas as pd
from sklearn.model_selection import train_test_split
```

```
reviews = pd.read_csv("/content/drive/MyDrive/NLP_Course/IMDB Dataset.csv")
reviews.head()
```

```
reviews.loc[reviews['sentiment']=='positive', 'sentiment'] = 1
reviews.loc[reviews['sentiment']=='negative', 'sentiment'] = 0
```

```
samples_review = reviews.sample(n=25000, random_state=42).reset_index(drop=True)
```

```
reviews, reviews_test = train_test_split(samples_review, test_size=0.08, stratify=samples_review.sentiment)
```

```
reviews_train, reviews_val = train_test_split(reviews, test_size=2000/23000, stratify=reviews.sentiment)
```

```
reviews_train.reset_index(drop=True).to_csv('/content/drive/MyDrive/NLP_Course/imdb_train.csv', index=False)
reviews_val.reset_index(drop=True).to_csv('/content/drive/MyDrive/NLP_Course/imdb_val.csv', index=False)
reviews_test.reset_index(drop=True).to_csv('/content/drive/MyDrive/NLP_Course/imdb_test.csv', index=False)
```
